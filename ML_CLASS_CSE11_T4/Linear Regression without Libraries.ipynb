```python
# -----------------------------------
# LINEAR REGRESSION (SUMMATION METHOD)
# -----------------------------------
# Your data
X = [11, 13, 4, 10, 11, 2]
Y = [2, 5, 8, 7, 3, 20]
n = len(X)

sum_x = 0
sum_y = 0
sum_xy = 0
sum_x2 = 0
for i in range(n):
    sum_x += X[i]
    sum_y += Y[i]
    sum_xy += X[i] * Y[i]
    sum_x2 += X[i] ** 2

#SLOPE (m)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - (sum_x ** 2))

#INTERCEPT (b)
b = (sum_y - m * sum_x) / n

#results
print("Slope (m):", m)
print("Intercept (b):", b)

#Predict
def predict(x):
    return m * x + b

print("Prediction for x = 15:", predict(15))
```

```python
# Linear Regression (mean method)

X = [11, 13, 4, 10, 11, 2]
Y = [2, 5, 8, 7, 3, 20]

n = len(X)

# Means
mean_x = sum(X) / n
mean_y = sum(Y) / n

# Slope (m)
num = sum((X[i] - mean_x) * (Y[i] - mean_y) for i in range(n))
den = sum((X[i] - mean_x) ** 2 for i in range(n))
m = num / den

# Intercept (b/c)
b = mean_y - m * mean_x

print("Slope (m):", m)
print("Intercept (b):", b)

def predict(x):
    return m * x + b

print("Prediction for x = 15:", predict(15))
```
