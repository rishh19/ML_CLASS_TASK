# Basic Neural Network

This is a simple neural network for binary classification.

The model consists of:

- Input layer
- Output layer (no hidden layer)

It uses sigmoid activation and gradient descent for training.

---

## Activation Function

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

---

## Model Equation

$$
y = \sigma(w \cdot x + b)
$$

---

## Loss Function

Mean Squared Error:

$$
L = \frac{1}{n} \sum (y - \hat{y})^2
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Simple dataset (linearly separable)
X = np.array([
    [1,2],
    [2,3],
    [3,3],
    [6,7],
    [7,8],
    [8,8]
])

y = np.array([[0],[0],[0],[1],[1],[1]])

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize weights
np.random.seed(0)
weights = np.random.randn(2,1)
bias = np.random.randn(1)

learning_rate = 0.1
epochs = 5000

loss_history = []

# Training
for epoch in range(epochs):
    
    # Forward
    z = np.dot(X, weights) + bias
    output = sigmoid(z)
    
    # Loss
    error = y - output
    loss = np.mean(error**2)
    loss_history.append(loss)
    
    # Backprop
    d_output = error * sigmoid_derivative(output)
    
    # Update
    weights += np.dot(X.T, d_output) * learning_rate
    bias += np.sum(d_output) * learning_rate

# Predictions
print("Predictions:")
print(np.round(output,3))

# Plot loss
plt.plot(loss_history)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
