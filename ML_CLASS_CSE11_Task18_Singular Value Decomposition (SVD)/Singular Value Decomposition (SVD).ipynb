Singular Value Decomposition (SVD)

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data matrix
A = np.array([
    [3,1,1],
    [-1,3,1]
])
```

```python
print("Original Matrix A:\n", A)

# Compute SVD
U, S, Vt = np.linalg.svd(A)

print("\nMatrix U:\n", U)
print("\nSingular Values:\n", S)
print("\nMatrix V^T:\n", Vt)
```

```python
# Convert singular values into diagonal matrix
Sigma = np.zeros((A.shape[0], A.shape[1]))
Sigma[:len(S), :len(S)] = np.diag(S)

print("\nSigma Matrix:\n", Sigma)
```

```python
# Reconstruct matrix
A_reconstructed = U @ Sigma @ Vt

print("\nReconstructed Matrix:\n", A_reconstructed)
```

```python
# Visualization
plt.imshow(A, cmap='viridis')
plt.title("Original Matrix")
plt.colorbar()
plt.show()

plt.imshow(A_reconstructed, cmap='viridis')
plt.title("Reconstructed Matrix using SVD")
plt.colorbar()
plt.show()
```
