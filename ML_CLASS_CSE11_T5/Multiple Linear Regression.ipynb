```python
n = int(input("enter the number of elements"))
x1=[]
x2=[]
y=[]

print("enter the x1 values:")
for i in range(n):
    x1.append(float(input()))

print("enter the x2 values:")
for i in range(n):
    x2.append(float(input()))

print("Enter y values:")
for i in range(n):
    y.append(float(input()))

#mean
x1_mean = sum(x1) / n
x2_mean = sum(x2) / n
y_mean = sum(y) / n

x1x1 = 0
x2x2 = 0
x1x2 = 0
x1y = 0
x2y = 0

for i in range(n):
    dx1 = x1[i] - x1_mean
    dx2 = x2[i] - x2_mean
    dy = y[i] - y_mean

    x1x1 += dx1 * dx1
    x2x2 += dx2 * dx2
    x1x2 += dx1 * dx2
    x1y  += dx1 * dy
    x2y  += dx2 * dy

# Coefficients
denominator = (x1x1 * x2x2) - (x1x2 * x1x2)

b1 = (x2x2 * x1y - x1x2 * x2y) / denominator
b2 = (x1x1 * x2y - x1x2 * x1y) / denominator
b0 = y_mean - b1 * x1_mean - b2 * x2_mean

print("b0 =", b0)
print("b1 =", b1)
print("b2 =", b2)
```
