```python
X = []
Y_actual = []

n_points = int(input("How many data points do you want to enter? "))

print(f"Please enter {n_points} pairs of X and Y values:")

for i in range(n_points):
    # Get input and convert to float (decimal number)
    x_val = float(input(f"Enter X for point {i+1}: "))
    y_val = float(input(f"Enter Y for point {i+1}: "))
    
    # Add them to our lists
    X.append(x_val)
    Y_actual.append(y_val)

print("\nData successfully loaded!")
print(f"X values: {X}")
print(f"Y values: {Y_actual}")
```

```python

#given
b_0 = 1.0  # Initial Intercept (Bias)
b_1 = 1.0  # Initial Slope (Weight)
alpha = 0.01  # Learning Rate

print(f"Initial Line: y = {b_0} + {b_1}x")

sum_error = 0
sum_error_x = 0

print("\n--- Calculation Step-by-Step ---")
print(f"{'X':<5} {'Y_act':<8} {'Y_pred':<8} {'Error':<8} {'Error*X':<8}")

n = len(X)
for i in range(n):
    # Calculate Prediction: y = b0 + b1*x
    y_pred = b_0 + (b_1 * X[i])
    
    # Calculate Error: (Y_pred - Y_actual)
    error = y_pred - Y_actual[i]
    
    # Calculate Error * X (for the slope gradient)
    error_x = error * X[i]
    
    # Accumulate the sums (Gradients)
    sum_error += error
    sum_error_x += error_x
    
    print(f"{X[i]:<5} {Y_actual[i]:<8} {y_pred:<8.2f} {error:<8.2f} {error_x:<8.2f}")

print("-------------------------------------")
print(f"Gradient for b_0 (Sum Error):    {sum_error}")
print(f"Gradient for b_1 (Sum Error * X): {sum_error_x}")

#Update the Parameters
#Formula: New = Old - Alpha * Gradient

new_b_0 = b_0 - (alpha * sum_error)
new_b_1 = b_1 - (alpha * sum_error_x)

#Final
print("\nFinal Result")
print(f"Updated Intercept (b_0): {new_b_0}")
print(f"Updated Slope (b_1):     {new_b_1}")
print(f"New Line Equation: y = {new_b_0:.2f} + {new_b_1:.2f}x")
```
