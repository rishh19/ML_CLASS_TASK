# Centroid Linkage Hierarchical Clustering

Centroid linkage is a hierarchical clustering method where the distance between two clusters is defined as the **distance between the centroids of the clusters**.

A centroid is the **mean of all data points in the cluster**.

---

## Centroid Formula

The centroid of a cluster is:

$$
C = \frac{1}{n}\sum_{i=1}^{n} x_i
$$

Where:

- \(C\) = centroid
- \(x_i\) = data points
- \(n\) = number of points in cluster

---

## Distance Between Clusters

Distance between clusters \(A\) and \(B\):

$$
D(A,B) = ||C_A - C_B||
$$

Where:

- \(C_A\) = centroid of cluster A
- \(C_B\) = centroid of cluster B

---

## Distance Metric

Usually Euclidean distance is used:

$$
d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Set random seed for reproducibility
np.random.seed(42)

# Sample dataset
X = np.array([
    [1,2],
    [2,3],
    [3,4],
    [5,7],
    [3,2],
    [8,7],
    [9,8],
    [7,6],
    [6,5]
])

# Create labels for each data point
labels = [f"P{i+1}" for i in range(len(X))]

# Perform hierarchical clustering using Centroid Linkage
Z = linkage(X, method='centroid')

# Plot dendrogram
plt.figure(figsize=(8,5))

dendrogram(
    Z,
    labels=labels,
    leaf_rotation=45,
    leaf_font_size=10
)

plt.title("Centroid Linkage Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")

plt.grid(True)
plt.tight_layout()
plt.show()
```

# Algorithm: Centroid Linkage Hierarchical Clustering

### Step 1: Start

Begin clustering process.

---

### Step 2: Input

Input dataset:

X = {x1, x2, ..., xn}

---

### Step 3: Initialize Clusters

Treat each data point as an individual cluster.

---

### Step 4: Compute Centroids

Compute centroid for each cluster.

---

### Step 5: Compute Distances

Compute distance between cluster centroids.

---

### Step 6: Merge Clusters

Merge the two clusters whose centroids have minimum distance.

---

### Step 7: Update Centroids

Compute new centroid of the merged cluster.

---

### Step 8: Repeat

Repeat Steps 5–7 until all data points are merged.
