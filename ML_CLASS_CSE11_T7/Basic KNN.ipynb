```python
import math

# Fixed 11 customers (Age, Loan)
X = [
    [22, 20000],
    [25, 25000],
    [30, 30000],
    [35, 35000],
    [40, 40000],
    [45, 45000],
    [50, 50000],
    [28, 27000],
    [33, 33000],
    [38, 38000],
    [48, 48000]
]
# Default status
Y = ['No','No','No','No','Yes','Yes','Yes','No','No','Yes','Yes']

# NEW customer
print("Enter NEW customer details")
age = int(input("Enter Age: "))
loan = int(input("Enter Loan Amount: "))
new = [age, loan]

k = int(input("Enter value of k: "))

# Step 1: Distance calculation
distances = []
for i in range(11):
    d = math.sqrt((new[0]-X[i][0])**2 + (new[1]-X[i][1])**2)
    distances.append([d, Y[i]])

# Step 2: Sort distances
distances.sort()

# Step 3: Take k nearest
nearest = distances[:k]
print("\nNearest neighbors:", nearest)

# Step 4: Voting
yes = 0
no = 0

for item in nearest:
    if item[1] == 'Yes':
        yes += 1
    else:
        no += 1

# Step 5: Prediction
if yes > no:
    print("\nPrediction: NEW customer WILL DEFAULT")
else:
    print("\nPrediction: NEW customer will NOT DEFAULT")
```
